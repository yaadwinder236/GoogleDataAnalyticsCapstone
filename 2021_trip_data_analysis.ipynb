The table has 13 columns and there are total 5595063 rows. The data type of columns started_at and ended_at is object which needs to be changed to datetime. No need to change datatype of rest of the columns.

```python
#the data types for start and end time of ride is object time, need to convert it to datetime
df[['started_at','ended_at']] = df[['started_at','ended_at']].astype('datetime64')
df.dtypes
```

```python
#clean string data, remove preceding or trailing extra spaces
df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
df.head()
```

There are a number of columns that have missing values. The rows containing missing values have to be removed. As ride_id uniquely identifies a ride, rows with duplicate ride_id also have to be removed.

```python
#remove duplicates based on ride_id
df.drop_duplicates(subset="ride_id", keep=False, inplace=True)
```

```python
#remove rows with missing values start_station, end_station, start/end lat/long
df2 = df.dropna()
df2.head()
```

```python
#remove rides tha lasted less than 1 min or more than 1 day (as suggested in project document)
#first create ride_length column by subtracting start datetime from end datetime
df2['ride_length'] = df2['ended_at'] - df2['started_at']
df2.head()
```

```python
#convert ride length to seconds
def timedelta_to_seconds(timedelta):
    return int(timedelta.total_seconds())
df2['ride_seconds'] = df2['ride_length'].apply(timedelta_to_seconds)
```

```python
df2.head()
```

```python
#remove rows where the ride length is less than 60 seconds or more than 86400 (24 hours) seconds
df3 = df2[(df2['ride_seconds']<86400) | (df2['ride_seconds']>60)]
df3.reset_index()
df3.head()
```

```python
#export data to csv for visualization in powerbi
df3.to_csv("cleaned_tripdata.csv",index=False)
```
